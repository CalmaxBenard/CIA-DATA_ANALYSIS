# Data Preparation and Model Discovery

```python
# Missing Data
df.isnull().sum()
```

```python
df[df["Agriculture"].isnull()]["Country"]
```

```python
df[df["Agriculture"].isnull()] = df[df["Agriculture"].isnull()].fillna(0)
```

```python
df[df["Agriculture"].isnull()]["Country"]
```

```python
df.isnull().sum()
```

```python
df["Climate"] = df.groupby("Region")["Climate"].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df["Literacy (%)"] = df.groupby("Region")["Literacy (%)"].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

## Data Feature Preapation

```python
# Drop country column (more os an index)
df = df.drop("Country", axis=1)
```

```python
# Dummies for the region column
X = pd.get_dummies(df)*1
```

```python
X
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

## Creating and Fitting KMeans Model

```python
from sklearn.cluster import KMeans
```

```python
# Choosing a k value
ssd = []

for k in range(2, 30):
    model = KMeans(n_clusters=k, n_init=15)
    model.fit(scaled_X)
    
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2, 30), ssd, "o--")
```

```python
pd.Series(ssd).diff().iloc[1:].plot(kind="bar")
```

```python
# Choosing K = 3
model = KMeans(n_clusters=3)
labels = model.fit_predict(scaled_X)
```

```python
labels
```

```python
X["K=3 Cluster Labels"] = labels
```

```python
X.corr()["K=3 Cluster Labels"].iloc[:-1].sort_values()
```

```python
plt.figure(figsize=(12,6), dpi=200)
X.corr()["K=3 Cluster Labels"].iloc[:-1].sort_values().plot(kind="bar")
```
